DESAFIO

```python
user = requests.get('https://api.github.com/users/GabrielSMarcal')
```

```python
status = user.status_code
dados = user.json()
url = user.url

print(f'Status: {status}')
print(f'Dados: {dados}')
print(f'URL: {url}')
```

```python
nome = dados['name']
nome_usuario = dados['login']
repositorios = dados['public_repos']
criacao = dados['created_at']

nome, nome_usuario, repositorios, criacao
```
